```python
import pandas as pd
import numpy as np 
from keras.models import Sequential  
from keras.layers import Dense 
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle 
from sklearn.decomposition import TruncatedSVD
from keras.callbacks import EarlyStopping
```

```python
filepath='E:/CRDC/运维工作表汇总.xlsx'
dataset = pd.read_excel(filepath)
dataset["分类"] = dataset["分类"].astype("category")
dataset["人员"] = dataset["人员"].astype("category")
pd.set_option('max_categories',27)
```

```python
for s in ['上海驻勤', '保密认证U盘处理', '内外网安全管理', '内外网服务器及应用运维', '内外网用户服务与管理', '内外网网络设备及应用运维', '基础应用管理', 
          '存储备份管理', '工作日夜班', '工作日白班', '布线调整', '广东分院驻勤', '应用运维', '故障处理', '机房管理', '深圳十四号线驻勤', '现场客服', '设备采购',
          '运维业务建设', '运维调度管理', '配线间管理', '铁总网络维护\桌面服务支持', '院内、铁总视频会议']:
    def contain(x,s):
        a = 0
        if x == s:
            a = a + 1
        return a 
    
    dataset['if_'+s]=dataset.分类.apply(lambda x: contain(x,s))
```

```python
for s in ['刘峰', '吕凤链', '姜书良', '孙先强', '孙博文', '孙惠彬', '张岩', '张律腾', '戴祝荔', '李开宇', '李晶', '李玲', '李萌', '洪莎莎', 
          '路浩', '韩寓', '马旺']:
    def contain(x,s):
        a = 0
        if x == s:
            a = a + 1
        return a 
    
    dataset['if_'+s]=dataset.人员.apply(lambda x: contain(x,s))
```

```python
def transfer(x):
    y = str(x)
    h = y[:2] 
    h_=int(h)
    m = y[3:5]
    m_=int(m)
    n = h_+m_/60
    k=n/21.5
    return k
dataset["起始"]=dataset["时间（起始）"].apply(lambda x :transfer(x))
```

```python
pd.set_option('max_columns',100)
data = dataset.values
svd = TruncatedSVD(n_components=20,algorithm='arpack')
svd_matrix = svd.fit_transform(data[:,9:51].tolist())
svd_feature = pd.DataFrame(svd_matrix)
svd_feature.columns = ['svd_'+str(i)for i in range(20)]
datacombine=pd.concat((dataset,svd_feature),axis=1)
```

```python
df=datacombine[datacombine['用时（分）']<= 210] 
df=shuffle(df)
data = df.values
```

```python
X_train, Y_train = data[:1500,9:50], data[:1500,3:4]       
X_test, Y_test = data[1500:,9:50], data[1500:,3:4]
X_train
```

```python
model = Sequential()  
model.add(Dense(40,input_dim=40, init = 'normal', activation = 'relu'))
model.add(Dense(20, init = 'normal', activation = 'relu'))
model.add(Dense(10, init = 'normal', activation = 'relu'))
model.add(Dense(1, init = 'normal'))
model.compile(loss = 'mse', optimizer = 'adam') 
```

```python
Earlystopping = EarlyStopping(monitor ='val_loss', patience = 2, verbose = 0, mode = 'min')
model.fit(X_train, Y_train, nb_epoch = 200,batch_size = 5,verbose = 2,callbacks = [Earlystopping])
```

```python
model.evaluate(X_test, Y_test, batch_size=5)  
```

```python
Y_pred = model.predict(X_test)     
```

```python
d1 = pd.DataFrame(Y_test,columns = ['实际观测值'])
d2 = pd.DataFrame(Y_pred,columns = ['预测值'])
d_result = pd.concat((d1, d2), axis = 1)
```

```python
from datetime import datetime
from datetime import timedelta
```

```python
def convert(x):
    y = str(x)
    z = y[-8:]
   
    if z != None:
        try:datetime.strptime(z,'%H:%M:%S') 
        except:pass
        return z
dataset["时间（完成）"]=dataset["时间（完成）"].apply(lambda x:convert(x))
```

```python
def convert(x):
    y = str(x)
    z = y[-8:]
   
    if z != None:
        try:datetime.strptime(z,'%H:%M:%S') 
        except:pass
        return z
dataset["时间（起始）"]=dataset["时间（起始）"].apply(lambda x:convert(x))
```

```python
def transfer(x):
    y = str(x)
    h = y[-8:-6] 
    h_=int(h)
    m = y[-5:-3]
    m_=int(m)
    n = (h_*60)+ m_
    return n
dataset["用时（分）"]=dataset["用时（分）"].apply(lambda x :transfer(x))
```

```python
dataset["小时数"]=dataset["用时（分）"]/60
```

```python
dataset.reset_index(drop=True,inplace=True)
```

```python
dataset['计数']=1
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties
myfont=FontProperties(fname=r'C:\Windows\Fonts\simhei.ttf',size=16)
sns.set(font=myfont.get_name())
```

```python
graph = df_group.pivot("分类","人员","计数")
f, ax = plt.subplots(figsize=(12,12))
sns.heatmap(graph, annot = True,fmt='.0f',linewidths=.5, ax=ax)
plt.show()
```

```python
f = plt.figure(figsize=(50,45))
sns.factorplot(x="人员",y="小时数",data=df_group2,kind='bar',size=7,aspect=1.6)

ax = plt.gca()

# ylim max value to be set
y_max = df_group2['小时数'].max() 
ax.set_ylim([0, y_max+10])

# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=14, color='black', ha='center', va='bottom')
  
plt.xticks(fontsize=14)   
plt.yticks(fontsize=14)            
plt.show()
```

```python
f = plt.figure(figsize=(12,10))
sns.factorplot(y="分类",x="小时数",data=df_group3,kind='bar',size=7,aspect=1.4)

ay = plt.gca()

# ylim max value to be set
y_max = df_group3['小时数'].max() 
ay.set_xlim([0, y_max+20])

# Iterate through the list of axes' patches
for p in ay.patches:
    ay.text( p.get_width()+15,p.get_y()+p.get_height(), '%d' % int(p.get_width()),
            fontsize=14, color='black', ha='center', va='bottom')
plt.xticks(fontsize=14)   
plt.yticks(fontsize=14)
plt.show()
```

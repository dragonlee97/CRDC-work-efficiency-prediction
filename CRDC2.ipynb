```python
df_group=dataset.groupby(["日期"])['用时（分）'].mean().reset_index()
```

```python
df_group.set_index("日期", inplace=True)
```

```python
df_group2=dataset.groupby(["日期"])['用时（分）'].sum().reset_index()
```

```python
df_group2.set_index("日期", inplace=True)
```

```python
df_group['用时（分）'].describe()
#outlier >208.5
```

```python
df_group2['用时（分）'].describe()
#outlier>7058.8
```

```python
df_mean=df_group[df_group['用时（分）']<=208.5] 
```

```python
df_sum=df_group2[df_group2['用时（分）']<=7058.8] 
```

```python
df_sum
```

```python
df_sum.to_csv('E:/CRDC/每日工作总时间.csv',index=True,encoding='utf-8')
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
rolmean = pd.rolling_mean(df_mean, window=5)
rolstd = pd.rolling_std(df_mean, window=5)
```

```python
f = plt.figure(figsize=(12,9))
orig = plt.plot(df_mean, color='blue',label='原始')
mean = plt.plot(rolmean, color='red', label='滚动平均')
std = plt.plot(rolstd, color='black', label = '滚动标准差')
plt.legend(loc='best')
plt.title('每日一项工作平均用时')
plt.show(block=False)
```

```python
rolmean2 = pd.rolling_mean(df_sum, window=5)
rolstd2 = pd.rolling_std(df_sum, window=5)
```

```python
f = plt.figure(figsize=(12,9))
orig = plt.plot(df_sum, color='blue',label='原始')
mean = plt.plot(rolmean2, color='red', label='滚动平均')
std = plt.plot(rolstd2, color='black', label = '滚动标准差')
plt.legend(loc='best')
plt.title('每日工作总用时')
plt.show(block=False)
```

```python
df=df_sum.squeeze()
```

```python
dftest =adfuller(df)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print (dfoutput)
#并不能拒绝原假设，非平稳时间序列
```
